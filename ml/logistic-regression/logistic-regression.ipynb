```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss, accuracy_score, precision_score, recall_score, f1_score

X = np.array([10, 2, 5, 3, 6, 8, 9, 13]).reshape(-1, 1) % 2
X_orig = np.array([10, 2, 5, 3, 6, 8, 9, 13])
y = np.array([1, 1, 0, 0, 1, 1, 0, 0])

model = LogisticRegression()

model.fit(X, y)

testing_set = np.array([[103], 
                        [102]]) % 2

y_pred = model.predict(testing_set)
y_prob = model.predict_proba(testing_set)[:,1]

y_answers = np.array([0, 1])
print("Result:", y_pred)
print("Probabilitis", y_prob)
print("Log Loss:", log_loss(y_answers, y_pred))
print("Accuracy Score", accuracy_score(y_answers, y_pred))
print("Precision", precision_score(y_answers, y_pred))
print("Recall:", recall_score(y_answers, y_pred))
print("F1:", f1_score(y_answers, y_pred))
```
