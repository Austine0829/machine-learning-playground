```python
from sklearn.linear_model import Lasso
import numpy as np

# Sample dataset
# 5 samples, 4 features
X = np.array([
    [1, 2, 3, 4],
    [2, 3, 4, 5],
    [3, 4, 5, 6],
    [4, 5, 6, 7],
    [5, 6, 7, 8]
])

# Target values (only first two features really affect y)
y = np.array([3, 5, 7, 9, 11])

# alpha = strength of regularization
lasso = Lasso(alpha=0.1)
lasso.fit(X, y)

# Check coefficients
print("Coefficients:", lasso.coef_)
print("Intercept:", lasso.intercept_)

# Keep only non-zero coefficient features
selected_features = X[:, lasso.coef_ != 0]

print("Selected features for next model:\n", selected_features)
```
