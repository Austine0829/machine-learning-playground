```python
from sklearn.linear_model import Ridge
import numpy as np

# Sample dataset
# 5 samples, 3 features
X = np.array([
    [1, 2, 1.9],
    [2, 4, 3.8],
    [3, 6, 5.7],
    [4, 8, 7.6],
    [5, 10, 9.5]
])

# Target values
y = np.array([3, 6, 9, 12, 15])

# alpha = regularization strength
ridge = Ridge(alpha=1.0)
ridge.fit(X, y)

# Check coefficients
print("Coefficients:", ridge.coef_)
print("Intercept:", ridge.intercept_)
```
